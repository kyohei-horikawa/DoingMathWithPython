```python
x,a,b,c,x0=symbols('x a b c x0')
```

```python
f=x**3-1
g=x**3+a*x**2+b*x+c
```

```python
x0=-1
df=f.diff()
df.subs(x,x0)
```

```python
f.subs(x,x0)-df.subs(x,x0)*x0
```

```python
# y=3x+1
# 3x-y+1=0
```

```python
d=abs(f.subs(x,x0)-df.subs(x,x0)*x0)/sqrt(df.subs(x,x0)**2+(-1)**2)
d
```

```python
dg=g.diff(x)
dg.subs(x,x0)-df.subs(x,x0)
b0=solve(dg.subs(x,x0)-df.subs(x,x0),b)[0]
b0
```

```python
# b=2a
```

```python
g.subs(x,x0).subs(b,b0)
c0=solve(g.subs(x,x0).subs(b,b0),c)[0]+f.subs(x,x0)
c0
```

```python
#c=-1+a
```

```python
a0=-2
```

```python
rev_g=g.subs(a,a0).subs(b,b0.subs(a,a0)).subs(c,c0.subs(a,a0))
rev_g
```

```python
rev_dg=rev_g.diff(x)
rev_dg
```

```python
x0,x1=solve(rev_dg)
x0,x1
```

```python
rev_g.subs(x,x0)
```

```python
rev_g.subs(x,x1)
```
